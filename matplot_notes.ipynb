 **Basic plotting**

1. Line plotting


````python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
plt.plot(a,b,)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(visible=True)
plt.title("Simple Line Plot")
plt.legend(["Line 1"])
plt.show()
````

**Explanation of each functions and sub-modules**

there are many sub-modules for matplotlib, but we are going through pyplot only for now.

pyplot is imported as plt for writing easily.

their are many functions which comes uder plt or pyplot.

1. plt.plot(x,y,fmt,**kargs)

this is plot function,  x and y represent data points of arrays or list 

**fmt - is called format string**

- Control color, marker, and line style with shorthand notation:

- Color: 'r' (red), 'b' (blue), 'g' (green), etc.

- Marker: 'o' (circles), 's' (squares), '.' (dots).

- Line: '-' (solid), '--' (dashed), ':' (dotted).

for example if we use 'ro--' it means red circled dashed line: note: it should be inside one 

mark string '', -- is **not** underscore.


````python
plt.plot(a,b,'ro--')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(visible=True)
plt.title("Simple Line Plot with Red Dots") 
plt.legend(["Line 1 with Red Dots"])
plt.show()
````

#### Keyword Arguments (**kwargs) ####

**Customize further:**

linewidth or lw: Line thickness.

markersize or ms: Marker size.

label: For legends.

alpha: Transparency (0 to 1).

````python
import numpy as np
import matplotlib.pyplot as plt


plt.plot([1,2,3,4],[5,6,7,8],'-b',markersize=2,linewidth=5,alpha=.5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(visible=True)
plt.show()
````

````python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'g-' ,label='sin(x)',linewidth=2) 
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() # this function is needed to show the label which is written in the plot function
plt.grid(True, linestyle=':', alpha=0.5) #look at the grid in : style with 50% transparency
plt.show()
````

**Grouped bar-chart**

multiple bars per category.

You use a grouped bar chart when:

	â€¢	You have data categorized by two categorical variables.
    
	â€¢	You want to compare subgroups within a main category.




````python
import numpy as np

# Data
labels = ['Group 1', 'Group 2', 'Group 3']
men = [29, 35, 30]
women = [25, 32, 34]

x = np.arange(len(labels))  # Label positions
width = .1  # Bar width

# Plot bars
plt.bar(x - width/2, men, width, label='Men', color='blue')
plt.bar(x + width/2, women, width, label='Women', color='pink')

# Customize
plt.title("Grouped Bar Chart")
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.xticks(x, labels)
plt.legend()

plt.show()
````

**Ploting using Histogram**


````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (1000 values, normally distributed)
data = np.random.randn(1000)  # Gaussian (normal) distribution

# Create a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Basic Histogram Example')

# Display the plot
plt.show()
````

````python
# Generate two datasets
data1 = np.random.normal(0, 1, 1000)  # Mean=0, Std=1
data2 = np.random.normal(3, 1.5, 1000)  # Mean=3, Std=1.5

# Plot both histograms
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1', color='blue',edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2', color='red',edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparing Two Distributions')
plt.legend()  # Show legend
plt.show()
````

f you use df['age'] directly, Plotly treats it as a Series, not a DataFrame, so you can't

Use df['age'] for matplotlib.

n Plotly Express histograms, the marginal argument adds extra plots to the histogram for 

additional context. The available types for marginal are:

"rug": Adds a rug plot (small vertical lines) showing individual data points.

"box": Adds a box plot summarizing the distribution.

"violin": Adds a violin plot showing the distribution's density.

"histogram": Adds a smaller histogram above or beside the main plot.

None: No marginal plot (default).

**output explanation**
Median: The line inside the box; it divides the data into two equal halves.

Q1 (First Quartile): The left (or bottom) edge of the box; 25% of data is below this value.

Q3 (Third Quartile): The right (or top) edge of the box; 75% of data is below this value.

Interquartile Range (IQR): The width of the box; calculated as Q3 - Q1.

Whiskers (Fences): Lines extending from the box to the smallest and largest values within 1.

5 Ã— IQR from Q1 and Q3.

Outliers: Points outside the whiskers.




````python
import plotly.express as px

# Scatter plot: Age vs Charges, colored by smoker status
fig = px.scatter(
    df,
    x='age',           # x-axis: age of customer
    y='charges',       # y-axis: medical charges
    color='smoker',    # color points by smoker status
    title='Age vs Charges by Smoker Status'
)
fig.show()
````

In Plotly Express, color='smoker' means the points are colored based on the values in the 

'smoker' column, automatically assigning different colors for each category (e.g., 'yes' or 

'no'). This helps visualize group differences.

In matplotlib, you cannot use color='smoker' because matplotlib expects a color name (like 

'red', 'blue') or a list/array of colors, not a column name. Matplotlib does not 

automatically map column values to colors.

Summary:

color='column' works in Plotly Express for automatic grouping.

In matplotlib, use explicit color names or arrays, not column names.
